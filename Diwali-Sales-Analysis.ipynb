##**Data Visuvalization...**

##**Gender...**

```python
# plotting a bar chart for Gender and it's count

a=sns.countplot(data=df,x="Gender",palette="Set1")


for bar in a.containers:
  a.bar_label(bar)
  plt.title("Gender Count")
  plt.tight_layout()

plt.show()
```

```python
# plotting a Pie chart for calculate the total sales as per the gender.

Sales_by_Gender=df.groupby("Gender")["Amount"].sum().sort_values(ascending=False)
explode=[0,0.1]

a=plt.pie(Sales_by_Gender,labels=Sales_by_Gender.index,autopct='%1.1f%%',shadow=True,explode=explode,startangle=90)

plt.title("Sales by Gender")
plt.legend()
plt.tight_layout()
plt.show()
```

*From above graphs we can see that most of the buyers are females and even the purchasing power of females are greater than men*

##**Age...**

```python
# plotting a bar chart for Age and it's count

plt.figure(figsize=(10,5))
a=sns.countplot(data=df,x="Age Group",palette="Set1",hue="Gender")

for bar in a.containers:
  a.bar_label(bar)
  plt.title("Age Group Count")
  plt.tight_layout()

plt.show()
```

```python
# plotting a bar chart for calculate the total sales as per the Age.

a=df.groupby("Age Group")["Amount"].sum().sort_values(ascending=False).reset_index()
plt.figure(figsize=(10,5))
a=sns.barplot(data=a,x="Age Group",y="Amount",palette="Set1")
plt.title("Total Sales by Age Group")
plt.tight_layout()


for bar in a.patches:
  plt.text(bar.get_x()+bar.get_width()/2,bar.get_height(),f'{bar.get_height():.1f}',ha="center",va="bottom",fontsize=10)
```

##***From above graphs we can see that most of the buyers are of age group between 26-35 yrs***

##**State**

```python
# plotting a bar chart for State and it's count

plt.figure(figsize=(10,5))
a=sns.countplot(data=df,x="State",palette="Set2",hue="Gender")
plt.title("State Count")
plt.xticks(rotation=90)
plt.tight_layout()

for bar in a.containers:
  a.bar_label(bar)
```

```python
# plotting a bar chart for calculate the total sales as per the Age.

ax=df.groupby("State")["Amount"].sum().sort_values(ascending=False).reset_index()
plt.figure(figsize=(10,5))
a=sns.barplot(data=ax,x="State",y="Amount",palette="Set3")
plt.title("Total Sales by State")
plt.xticks(rotation=90)
plt.tight_layout()

for bar in a.patches:
  plt.text(bar.get_x()+bar.get_width()/2,bar.get_height(),f'{bar.get_height():.1f}',ha="center",va="bottom",fontsize=7)
```

*From above graphs we can see that most of the orders & total sales/amount are from Uttar Pradesh, Maharashtra and Karnataka respectively*


##**Marital Status**

```python
# plotting a bar chart for Marital Status and it's count
a=sns.countplot(data=df,x="Marital_Status",palette="Set2")
plt.title("Marital Status Count")
plt.tight_layout()

for bar in a.containers:
  a.bar_label(bar)
```

```python
# plotting a bar chart for calculate the total sales as per the Marital Status and Gender.

a=df.groupby(["Marital_Status","Gender"])["Amount"].sum().reset_index()
plt.figure(figsize=(10,5))
a=sns.barplot(data=a,x="Marital_Status",y="Amount",palette="Set2",hue="Gender")
plt.title("Total Sales by Marital Status")

for bar in a.patches:
  plt.text(bar.get_x()+bar.get_width()/2,bar.get_height(),f'{bar.get_height():.1f}',ha="center",va="bottom",fontsize=10)

plt.show()
```

*From above graphs we can see that most of the buyers are married (women) and they have high purchasing power*

##**Ocupation**

```python
# Plotting a bar chart for Occupation and it's count

plt.figure(figsize=(10,5))
ax=sns.countplot(data=df,x="Occupation",palette="Set2")
plt.title("Occupation Count")
plt.xticks(rotation=90)
plt.tight_layout()

for bar in ax.containers:
  ax.bar_label(bar)
```

```python
# plotting a bar chart for calculate the total sales as per the Occupation.

ax=df.groupby("Occupation")["Amount"].sum().sort_values(ascending=False).reset_index()
plt.figure(figsize=(12,5))
a=sns.barplot(data=ax,x="Occupation",y="Amount",palette="Set2")
plt.title("Total Sales by Occupation")
plt.xticks(rotation=90)
plt.tight_layout()

for bar in a.patches:
  plt.text(bar.get_x()+bar.get_width()/2,bar.get_height(),f'{bar.get_height():.1f}',ha="center",va="bottom",fontsize=7)
```

*From above graphs we can see that most of the buyers are working in IT, Healthcare and Aviation sector*

##**Product Category**

```python
# Plotting a bar chart for Product Category and it's count

plt.figure(figsize=(10,5))
ax=sns.countplot(data=df,x="Product_Category",palette="Set1")
plt.title("Product Category Count")
plt.xticks(rotation=90)
plt.tight_layout()

for bar in ax.containers:
  ax.bar_label(bar)
plt.show()
```

```python
# plotting a bar chart for calculate the top 10 total sales as per the Product Catagory.

ax=df.groupby("Product_Category")["Amount"].sum().sort_values(ascending=False).reset_index().head(10)
plt.figure(figsize=(10,5))
a=sns.barplot(data=ax,x="Product_Category",y="Amount",palette="Set1")
plt.xticks(rotation=90)
plt.title("Total Sales by Product Category")
plt.tight_layout()

for bar in a.patches:
  plt.text(bar.get_x()+bar.get_width()/2,bar.get_height(),f'{bar.get_height():.1f}',ha="center",va="bottom",fontsize=8)
```

*From above graphs we can see that most of the sold products are from Food, Clothing and Electronics category*

```python
# top 10 most sold products

ax=df.groupby("Product_ID")["Orders"].sum().sort_values(ascending=False).reset_index().head(10)
plt.figure(figsize=(13,5))
a=sns.barplot(data=ax,x="Product_ID",y="Orders",palette="Set1")
plt.title("Top 10 Most Sold Products")
plt.xlabel("Product ID")
plt.ylabel("Number of Orders")
plt.tight_layout()

for bar in a.patches:
  plt.text(bar.get_x()+bar.get_width()/2,bar.get_height(),f'{bar.get_height():.0f}',ha="center",va="bottom",fontsize=10)
```

##**Summary...**

**2.Females aged 26–35 lead in shopping activity.**
